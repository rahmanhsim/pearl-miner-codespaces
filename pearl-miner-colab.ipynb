```
# Start
import subprocess

cmd = f"./pearl-miner --host {POOL} --user {WALLET} --worker {WORKER}"
print(f"Starting: {cmd}")

process = subprocess.Popen(
    cmd, shell=True,
    stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True
)

for line in process.stdout:
    print(line.rstrip())
```
